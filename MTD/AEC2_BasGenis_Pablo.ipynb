```python
# PROBLEM 1

"""
We are told that the waiting times in an emergency department are normally distributed.
We took a sample of 100 people and found that the average waiting time was 14.25 minutes
with a standard deviation of 2.5 minutes. We want to test whether this differs significantly
from the known value of 15 minutes.

We'll use the Z-test since the sample size is large and the population standard deviation is known.
"""

from math import sqrt

# Given values
mu_0 = 15            # Hypothesized population mean
x_bar = 14.25        # Sample mean
sigma = 2.5          # Standard deviation
n = 100              # Sample size

# a) Test at 5% significance level
z_alpha_2 = 1.96
z_score = (x_bar - mu_0) / (sigma / sqrt(n))

print("Problem 1.a:")
print("Z-score:", z_score)
if abs(z_score) > z_alpha_2:
    print("Since the Z-score is greater than 1.96, we reject the null hypothesis. So, we can say that the average waiting time is significantly different from 15 minutes at the 5% level.")
else:
    print("Since the Z-score is not greater than 1.96, we fail to reject the null hypothesis. So, we can't say there's a significant difference.")

# b) Test at 0.1% significance level
z_alpha_2_001 = 3.27

print("\nProblem 1.b:")
if abs(z_score) > z_alpha_2_001:
    print("Z-score is less than 3.27, so we fail to reject the null hypothesis. There's not enough evidence to say the average is different at the 0.1% level.")
else:
    print("Same as above: we fail to reject the null hypothesis.")

# c) Is there a contradiction?
print("\nProblem 1.c:")
print("No contradiction. At 5%, the difference is considered significant, but at the much stricter 0.1% level, it's not. This just shows that our evidence isn't extremely strong.")
```

```python
# PROBLEM 2

"""
We want to study how the number of pages of a job (X) relates to the printing time (Y).
We're given summary statistics and regression output, and we need to interpret them.
"""

# a) Estimate parameters
print("\nProblem 2.a:")
print("The regression equation is: Y = 2.6272 + 0.428601 * X. This means the estimated printing time increases by about 0.43 minutes for each extra page.")

# b) Goodness of fit (R^2)
SSR = 39343.3
SST = 47368.9
R2 = SSR / SST
print("\nProblem 2.b:")
print(f"R-squared = {R2:.4f}, which means that around {R2*100:.2f}% of the variability in printing time is explained by the number of pages.")

# c) Correlation coefficient
Var_X = 8.08757
Var_Y = 640.121
Cov_XY = 65.5735

r = Cov_XY / sqrt(Var_X * Var_Y)
print("\nProblem 2.c:")
print(f"The correlation coefficient is {r:.4f}. This shows a strong positive linear relationship between number of pages and printing time.")

# d) Statistical significance
print("\nProblem 2.d:")
print("Looking at the output, both the constant and the slope have very small p-values (close to 0), so they are statistically significant at α = 0.05. The final model is: Y = 2.6272 + 0.428601 * X")

# e) Predict for 6 pages
X_new = 6
Y_pred = 2.6272 + 0.428601 * X_new
print("\nProblem 2.e:")
print(f"If a job has 6 pages, the expected printing time is approximately {Y_pred:.2f} minutes.")

# f) Residual variance
Residual_SS = 8025.61
Residual_df = 73
residual_variance = Residual_SS / Residual_df
print("\nProblem 2.f:")
print(f"The residual variance is {residual_variance:.2f}. This tells us the average squared error between the observed and predicted printing times.")
```
